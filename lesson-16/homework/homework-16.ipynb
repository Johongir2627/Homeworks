```python
#1. Convert List to 1D Array
import numpy as np
a = [1, 2, 3, 4, 5]
a = np.array(a)
print(a)
```

```python
#2 Create 3x3 Matrix (2?10)
import numpy as np
a = np.arange(2, 11).reshape(3, 3)
print(a)
```

```python
#3. Null Vector (10) & Update Sixth Value
a = np.zeros(10)
a[5] = 11
print(a)
```

```python
#4. Array from 12 to 38
array = np.arange(12, 39)
print(array)
```

```python
#5 Conver array to float type
a = np.array([1, 2, 3, 4])
a = a.astype(float)
print(a)
print(a.dtype)
```

```python
#6 . Celsius to Fahrenheit Conversion
c = np.array([0, 12, 45.21, 34, 99.91])
f = (c * 9/5) + 32
print(f)
```

```python
#7. Append values to array
a = np.array([1, 2, 3])
a = np.append(a, [4, 5, 6])
print(a)
```

```python
#8. Array statistics functions
a = np.random.rand(10)
mean = np.mean(a)
median = np.median(a)
std = np.std(a)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

```

```python
#9. find min and max values
a = np.random.rand(10)
min_value = np.min(a)
max_value = np.max(a)
print("Min:", min_value)
print("Max:", max_value)
```

```python
#10 Create a 3x3x3 array with random values.
a = np.random.rand(3, 3, 3)
print(a)
print(a.shape)
```
